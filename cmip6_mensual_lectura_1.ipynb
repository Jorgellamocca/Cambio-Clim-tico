```python
import intake

cat_url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"
cat = intake.open_esm_datastore(cat_url)
df = cat.df

# Filtro principal
df_sel = df[
    (df["variable_id"].isin(["pr", "tas", "tasmax", "tasmin"])) &
    (df["experiment_id"].isin(["historical", "ssp126", "ssp245", "ssp370", "ssp585"]))
]

# Listas finales
modelos = sorted(df_sel["source_id"].unique())
variables = sorted(df_sel["variable_id"].unique())
escenarios = sorted(df_sel["experiment_id"].unique())
grillas = sorted(df_sel["grid_label"].unique())

print("\nModelos CMIP6:")
for m in modelos:
    print("-", m)

print("\nVariables:")
for v in variables:
    print("-", v)

print("\nEscenarios:")
for e in escenarios:
    print("-", e)

print("\nTipo de grilla:")
for g in grillas:
    print("-", g)
```

```python
# --------------------------------------------------
# Selección
# --------------------------------------------------
MODELOS = [
    "ACCESS-ESM1-5",
    "CMCC-ESM2",
    "CNRM-CM6-1",
    "MRI-ESM2-0",
    "NorESM2-MM",
]

VARIABLES = ["pr", "tas", "tasmax", "tasmin"]
ESCENARIOS = ["historical", "ssp126", "ssp245", "ssp370", "ssp585"]

df_sel = df[
    (df["source_id"].isin(MODELOS)) &
    (df["variable_id"].isin(VARIABLES)) &
    (df["experiment_id"].isin(ESCENARIOS))
]

# --------------------------------------------------
# Construir tabla resumen (con Institución)
# --------------------------------------------------
tabla = (
    df_sel.groupby("source_id")
    .agg(
        Institucion=("institution_id", lambda x: ", ".join(sorted(x.unique()))),
        Variables=("variable_id", lambda x: ", ".join(sorted(x.unique()))),
        Escenarios=("experiment_id", lambda x: ", ".join(sorted(x.unique()))),
        Grilla=("grid_label", lambda x: ", ".join(sorted(x.unique())))
    )
    .reset_index()
    .rename(columns={"source_id": "Modelo"})
)

tabla
```
