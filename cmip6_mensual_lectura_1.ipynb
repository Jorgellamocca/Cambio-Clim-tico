```python
import intake
import pandas as pd

# --------------------------------------------------
# Abrir cat谩logo CMIP6
# --------------------------------------------------
cat_url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"
cat = intake.open_esm_datastore(cat_url)
df = cat.df

# --------------------------------------------------
# Selecci贸n
# --------------------------------------------------
MODELOS = [
    "ACCESS-ESM1-5",
    "CMCC-ESM2",
    "CNRM-CM6-1",
    "MRI-ESM2-0",
    "NorESM2-MM",
]

VARIABLES = ["pr", "tas", "tasmax", "tasmin"]
ESCENARIOS = ["historical", "ssp126", "ssp245", "ssp370", "ssp585"]

df_sel = df[
    (df["source_id"].isin(MODELOS)) &
    (df["variable_id"].isin(VARIABLES)) &
    (df["experiment_id"].isin(ESCENARIOS))
]

# --------------------------------------------------
# Funci贸n: miembros por escenario
# --------------------------------------------------
def miembros_por_escenario(df_modelo, escenario):
    miembros = (
        df_modelo[df_modelo["experiment_id"] == escenario]["member_id"]
        .dropna()
        .unique()
    )
    return ", ".join(sorted(miembros)) if len(miembros) > 0 else "-"

# --------------------------------------------------
# Construir tabla final
# --------------------------------------------------
filas = []

for modelo, df_m in df_sel.groupby("source_id"):
    fila = {
        "Modelo": modelo,
        "Institucion": ", ".join(sorted(df_m["institution_id"].dropna().unique())),
        "Variables": ", ".join(sorted(df_m["variable_id"].unique())),
        "Grilla": ", ".join(sorted(df_m["grid_label"].unique())),
    }

    for esc in ESCENARIOS:
        fila[f"Miembros_{esc}"] = miembros_por_escenario(df_m, esc)

    filas.append(fila)

tabla_final = pd.DataFrame(filas)

tabla_final
```

```python
# ==================================================
# 3. Funci贸n: elegir miembro hist贸rico de referencia
# ==================================================
def miembro_historico_referencia(df_sel):
    miembros = (
        df_sel[df_sel["experiment_id"] == "historical"]["member_id"]
        .dropna()
        .unique()
    )

    if len(miembros) == 0:
        return "-"

    # Prioridad absoluta: r1i1p1f*
    r1_ref = [m for m in miembros if re.match(r"^r1i1p1f\d+$", m)]
    if r1_ref:
        return r1_ref[0]

    # Luego cualquier r1*
    r1 = [m for m in miembros if m.startswith("r1")]
    if r1:
        return r1[0]

    # Finalmente menor r<num> (orden num茅rico)
    return sorted(
        miembros,
        key=lambda x: int(re.findall(r"^r(\d+)", x)[0])
    )[0]

# ==================================================
# 4. Construir TABLA FINAL (100% consistente)
# ==================================================
filas = []

for modelo, df_m in df_sel.groupby("source_id"):

    #  miembro hist贸rico de referencia
    ref_hist = miembro_historico_referencia(df_m)

    fila = {
        "Modelo": modelo,
        "Institucion": ", ".join(sorted(df_m["institution_id"].dropna().unique())),
        "Variables": ", ".join(sorted(df_m["variable_id"].unique())),
        "Grilla": ", ".join(sorted(df_m["grid_label"].unique())),
    }

    # Usar el MISMO miembro en todos los escenarios
    for esc in ESCENARIOS:
        miembros_esc = (
            df_m[df_m["experiment_id"] == esc]["member_id"]
            .dropna()
            .unique()
        )

        fila[f"Miembro_{esc}"] = ref_hist if ref_hist in miembros_esc else "-"

    filas.append(fila)

tabla_final = pd.DataFrame(filas)

tabla_final
```
